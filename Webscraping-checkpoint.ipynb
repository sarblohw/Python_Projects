```python
all_tables = soup.find_all('table')
all_tables[5]
```

```python
rows = all_tables[5].find_all('tr')
```

```python
df = pd.DataFrame(columns = ['Rank', 'Country', 'Population', 'Area', 'Density'])

for i, row in enumerate(rows):
    cols = row.find_all('td')
    if cols != []:
        rank = cols[0].text
        name = cols[1].text
        population = cols[2].text.strip()
        area = cols[3].text.strip()
        density = cols[4].text.strip()
        df = df.append({"Rank" : rank, "Country" : name, "Population" : population, "Area" : area, "Density" : density}, ignore_index = True)
df
```
