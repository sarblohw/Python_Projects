```python
df_best.columns = ['', 'Budget', 'Revenue', 'Number of Votes', 'Avg.Voting Score', 'Popularity', 'Profit', 'ROI']
```

```python
df_best.loc[df_best.Budget >= 5].sort_values(['ROI','Budget'], ascending=[False,False])
```

```python
df_best['Number of Votes'].fillna(value = 0, inplace=True)
```

```python
df_best.isnull().sum()
```

```python
def best_worst_movies(n, by, ascending=False, min_budget =0, min_votes=0):
    result_df = df_best.loc[(df_best['Budget'] >= min_budget) & (df_best['Number of Votes'] >= min_votes), ['', df_best[by]]
                            .sort_values(by = by, ascending = ascending).head(n).copy()
    return HTML(result_df.to_html (escape=False))
```

```python
best_worst_movies(10, 'Revenue', ascending = False, min_budget = 3, min_votes = 3)
```

```python
best_worst_movies(5, 'ROI')
```
